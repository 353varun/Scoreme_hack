```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# Example usage
pdf_path = "test3.pdf"  # Replace with your PDF file path
extracted_text = extract_text_from_pdf(pdf_path)

print(extracted_text)
```

```python
from langchain_core.prompts import  PromptTemplate

prompt_extract = PromptTemplate.from_template(
        """"
        ### SCRAPED TEXT FROM PDFs:
        {extracted_text}
        ### INSTRUCTION:
        You are an expert in text processing and table extraction. Your task is to accurately identify, structure, and extract tables from raw text obtained from a system-generated PDF. 
        The tables may have borders, no borders, or irregular shapes. Ensure that the extracted tables maintain the correct row and column structure, handling cases such as merged cells
         and multi-line content. Output the structured table data in a JSON format where each table is a list of rows, with each row containing a list of column values. Maintain data
          integrity and formatting. 
        ### DONOT USE TABULA OR CAMELOT LIBRARY.
        #### VALID JSSON(NO PREAMBLE):
        """
)

chain_extract = prompt_extract|llm
res=chain_extract.invoke(input={'extracted_text':extracted_text})
json_output=res.content
```

```python
import json
import pandas as pd

# Example JSON response from LLM
json_data = res.content

# Load JSON data
data = json.loads(json_data)

# Create an Excel writer object
excel_filename = "extracted_tables.xlsx"
with pd.ExcelWriter(excel_filename, engine="openpyxl") as writer:
    for table in data["tables"]:
        df = pd.DataFrame(table["rows"], columns=table["headers"])
        
        # Write each table to a separate sheet in the same Excel file
        sheet_name = f"Table_{table['table_id']}"
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"Excel file '{excel_filename}' created successfully!")
```
